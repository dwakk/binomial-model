# Modèle binomial

__Objectif:__ Construire un modle binomial de Cox-Ross-Rubinstein (CRR) pour évaluer le prix d'une option financière. À chaque intervalle de temps $\Delta t$, le prix peut monter d'un facteur _u_ ou descendre d'un facteur _d_


#### Paramètres du modèle


On définit les paramètres d'entrée de l'arbre:

- $S0$: prix initial
- $t$: durée en années
- $\sigma$: volatilité
- $n$: taille de l'arbre

On peut maintenand calculer _u_ et _d_:

$u = e^{\sigma \sqrt{t}}$
<br>
$d = \frac{1}{u}$


#### Création de l'arbre

On crée une structure récursive de noeud 

```python
class Node():
    def __init__(self, price, depth=0):
        self.price = price
        self.depth = depth
        self.up = None
        self.down = None

S0 = 100

root = Node(S0)
root.up = Node(S0 * 1.1, root.depth + 1)
root.down = Node(S0 * 0.9, root.depth + 1)

print(f'Valeur si up: {root.up.price}')
print(f'Valeur si down: {root.down.price}')
```
